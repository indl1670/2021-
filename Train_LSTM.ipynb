# Train LSTM

```python
from EQTransformer.core.trainer import trainer
trainer(input_hdf5='/home/STED/new_chunk1.hdf5',
        input_csv='/home/STED/new_chunk1.csv',
        output_name='chunk1_LSTM',                
        cnn_blocks=2,
        lstm_blocks=1,
        padding='same',
        activation='relu',
        drop_rate=0.2,
        label_type='gaussian',
        add_event_r=0.6,
        add_gap_r=0.2,
        shift_event_r=0.9,
        add_noise_r=0.5, 
        mode='generator',
        train_valid_test_split=[0.8, 0.1, 0.1],
        batch_size=20,
        epochs=5, 
        patience=2,
        gpuid=0,
        gpu_limit=1)
```
